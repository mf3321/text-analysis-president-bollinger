```python
# Importing libraries

import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
```

```python
# Accessing text files

files = ["2003_Bollinger.txt",
        "2004_Bollinger.txt",
        "2005_Bollinger.txt",
        "2006_Bollinger.txt",
        "2007_Bollinger.txt",
        "2008_Bollinger.txt",
        "2009_Bollinger.txt",
        "2010_Bollinger.txt",
        "2011_Bollinger.txt",
        "2012_Bollinger.txt",
        "2013_Bollinger.txt",
        "2014_Bollinger.txt",
        "2015_Bollinger.txt",
        "2016_Bollinger.txt",
        "2017_Bollinger.txt",
        "2018_Bollinger.txt",
        "2019_Bollinger.txt",
        "2020_Bollinger.txt",
        "2021_Bollinger.txt",
        "2022_Bollinger.txt",
        "2023_Bollinger.txt"]

# Creating empty list

paperwords = []

# Using readlines method to return a list containing each line in each file 
# as a list item

for d in files:
    with open( str(d), encoding = 'latin1') as nf:
        lines = nf.readlines()        

# Using split method to split lines into a list of individual words

    i = 1
    for x in lines: 
        words = x.split()
        
# Removing punctuation attached to individual words; appending file name, 
# line, and individual word as string to empty list    
        for w in words:
            w = w.strip('''!()-[]{};:'"\,<>./?@#$%^&*_~''')
            paperwords.append([d, i, w])
            
        i += 1
```

```python
# Checking paperwords list created as intended

paperwords[:30]
```

```python
# Obtaining total number of words in all 21 commencement addresses including
# stopwords

len(paperwords)
```

```python
# Importing stop words from Natural Language Toolkit (NLTK) corpus

from nltk.corpus import stopwords
```

```python
# Reviewing English-language stop words from NLTK corpus

print(stopwords.words('english'))
```

```python
# Creating a separate list with NLTK English-language stop words

stop_words = list(stopwords.words('english'))
```

```python
# # Checking stop_words list created as intended

stop_words[:30]
```

```python
# Removing stop words from paperwords

for word in paperwords:
  if word[2] in stop_words: 
    paperwords.remove(word)
    
paperwords[:30]
```

```python
# Obtaining total number of words in all 21 commencement addresses less
# stopwords

len(paperwords)
```

```python
# Creating a dataframe

ppwords = pd.DataFrame(paperwords, columns = ['speech', 'line', 'word'])
ppwords
```

```python
# Accessing sentiment dataframe

df_bing = pd.read_csv('C:/Users/May/Desktop/lab2/Python Notebooks for Lab/bing.csv')
df_bing.head()
```

```python
# Qualifying words as either positive or negative sentiment; inner merging on 
# word

result = ppwords.merge(df_bing, on = 'word')
result.head(-5)
```

```python
# Relabeling positive sentiment as 1 and negative sentiment as -1
result['sentiment'] = result['sentiment'].replace('positive', 1)
result['sentiment'] = result['sentiment'].replace('negative', -1)
result.head(-5)
```

```python
# Creating a percentage column to identify progress in Bollinger's address

result['percentage'] = round(result['line'] / max(result['line']) *100/5)*5
result.head(-5)
```
