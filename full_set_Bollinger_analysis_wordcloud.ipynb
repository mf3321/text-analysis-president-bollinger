```python
# Importing libraries

import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
```

```python
# Accessing text files

files = ["2003_Bollinger.txt",
        "2004_Bollinger.txt",
        "2005_Bollinger.txt",
        "2006_Bollinger.txt",
        "2007_Bollinger.txt",
        "2008_Bollinger.txt",
        "2009_Bollinger.txt",
        "2010_Bollinger.txt",
        "2011_Bollinger.txt",
        "2012_Bollinger.txt",
        "2013_Bollinger.txt",
        "2014_Bollinger.txt",
        "2015_Bollinger.txt",
        "2016_Bollinger.txt",
        "2017_Bollinger.txt",
        "2018_Bollinger.txt",
        "2019_Bollinger.txt",
        "2020_Bollinger.txt",
        "2021_Bollinger.txt",
        "2022_Bollinger.txt",
        "2023_Bollinger.txt"]

# Creating empty list

bagowords = []

# Using readlines method to return a list containing each line in each file 
# as a list item

for d in files:
    with open( str(d), encoding = 'latin1') as nf:
        lines = nf.readlines()        

# Using split method to split text into a list of individual words

    for x in lines: 
        words = x.split()
        
# Removing punctuation attached to individual words; appending file name, 
# line, and the individual word as string to empty list      
        for w in words:
            w = w.strip('''!()-[]{};:'"\,<>./?@#$%^&*_~''')
            bagowords.append(w)
```

```python
# Checking bagowords list created as intended

bagowords[:30]
```

```python
# Obtaining total number of words in all 21 commencement addresses including
# stopwords

len(bagowords)
```

```python
# Importing stop words from Natural Language Toolkit (NLTK) corpus

from nltk.corpus import stopwords
```

```python
# Reviewing English-language stop words from NLTK corpus

print(stopwords.words('english'))
```

```python
# Creating a separate list with NLTK English-language stop words

stop_words = list(stopwords.words('english'))
```

```python
# Checking stop_words list created as intended

stop_words[:30]
```

```python
# Removing stop words from bagowords

for word in bagowords:
  if word in stop_words: 
    bagowords.remove(word)
    
bagowords[:30]
```

```python
# Obtaining total number of words in all 19 commencement addresses, less stop
# words

len(bagowords)
```

```python
# Creating a wordcloud for all 19 commencement addresses

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

cloudwords = " ".join(bagowords)
Bollinger_wordcloud = WordCloud(background_color='white').generate(cloudwords)

plt.figure(figsize=(10,7))
plt.imshow(Bollinger_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
