```python
# Importing libraries

import pandas as pd
import numpy as np
import nltk
```

```python
# Accessing text files

files = ["2003_Bollinger.txt",
        "2004_Bollinger.txt",
        "2005_Bollinger.txt",
        "2006_Bollinger.txt",
        "2007_Bollinger.txt",
        "2008_Bollinger.txt",
        "2009_Bollinger.txt",
        "2010_Bollinger.txt",
        "2011_Bollinger.txt",
        "2012_Bollinger.txt",
        "2013_Bollinger.txt",
        "2014_Bollinger.txt",
        "2015_Bollinger.txt",
        "2016_Bollinger.txt",
        "2017_Bollinger.txt",
        "2018_Bollinger.txt",
        "2019_Bollinger.txt",
        "2020_Bollinger.txt",
        "2021_Bollinger.txt",
        "2022_Bollinger.txt",
        "2023_Bollinger.txt"]

paperwords = []

# Using readlines method to return a list containing each line in each file 
# as a list item

for d in files:
    with open( str(d), encoding = 'latin1') as nf:
        lines = nf.readlines()        

# Using split method to split text into a list of individual words

    for x in lines: 
        words = x.split()
        
# Removing punctuation attached to individual words; appending file name, 
# line, and individual word as string to empty list       
        for w in words:
            w = w.strip('''!()-[]{};:'"\,<>./?@#$%^&*_~''')
            paperwords.append([d, w])
```

```python
# Importing stop words from Natural Language Toolkit (NLTK) corpus

from nltk.corpus import stopwords
```

```python
# Reviewing English-language stop words from NLTK corpus

print(stopwords.words('english'))
```

```python
# Creating a separate list with NLTK English-language stop words

stop_words = list(stopwords.words('english'))
```

```python
# # Checking stop_words list created as intended

stop_words[:30]
```

```python
# Removing stop words from paperwords

for word in paperwords:
  if word[1] in stop_words: 
    paperwords.remove(word)
    
paperwords[:30]
```

```python
# Creating a dataframe

ppwords = pd.DataFrame(paperwords, columns = ['speech', 'word'])
ppwords
```

```python
# Creating another dataframe

d = ppwords.groupby(['speech', 'word']).size().reset_index(name='n')
d
```

```python
# Creating a CountVectorizer object and fit it on the documents

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
dtm = vectorizer.fit_transform(d['word'])
```

```python
# Creating yet another dataframe

dtm_df = pd.DataFrame(dtm.todense(),
                      index=d['speech'],
                      columns=vectorizer.get_feature_names())
dtm_df
```

```python
# Creating one last dataframe

new_df = dtm_df.groupby(dtm_df.index).sum()
new_df
```
